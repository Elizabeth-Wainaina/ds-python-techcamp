```python
x = petrol_df.iloc[:,:-1]
x
```

```python
y = petrol_df["petrol_consumption"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.55, random_state=0)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, x.columns, columns=['Coefficient'])
coeff_df
```

y(petrol_consumption) = 1341.86 x1(population_driver_licence) -40.01 x2(petrol_tax) - 0.065 x3(average_income) - 0.0047 x4(paved_highways)

```python
regressor.intercept_
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
results_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

```python
results_df
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
