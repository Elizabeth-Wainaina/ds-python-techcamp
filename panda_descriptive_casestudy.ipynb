### Describe the data graphically 

#### 1. Box and whisker plot

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.set_ylabel('school_rating')
sr_rl.boxplot(by='school_rating', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

### 2. Scatter plot 

```python
plt.figure(figsize=(14,8)) # set the size of the graph
sr_rl.regplot(data=df, x='reduced_lunch', y='school_rating')
```
