```python
gn_ru = df_movies_one[['genre', 'reviews_from_users']]
gn_ru.head(10)
```

```python
# any row that has teh genre drama will be replaced by the 6.6
# 4.0+9.0+7.0 = 20/3 = 6.6
```

```python
gn_ru = df_movies_one[['genre', 'reviews_from_users']]
(gn_ru['genre'] == 'Drama', 'History')
gn_ru[(gn_ru['genre'] == 'Drama, History')]
```

```python
gns =list(set(gn_ru['genre']))
gns
```

```python
# the indexing 0 represents the first genre displayed
gnr = gn_ru[gn_ru["genre"]==gns[0]]
gnr
```

```python
type(gnr)
```

```python
gnr.mean()
```

```python
gnm = {}
```

```python
for i in range(0, len(gns)):
    gnr = gn_ru[gn_ru["genre"]==gns[i]]
    gnm[gns[i]]=list(gnr.mean()) [0]
```
